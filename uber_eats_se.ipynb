```python
df_stores=pd.read_csv('supermercados_ubereats_limpio.csv')
df_stores
```

```python
productosdf=pd.read_excel("prod_pacic.xlsx")
productosdf.columns=productosdf.columns.str.lower()
#Eliminar espacios al frente y atr√°s
productosdf['producto']=productosdf['producto'].str.strip()
productosdf['producto']=productosdf['producto'].str.lower()
#Eliminar acentos
productosdf['producto']=productosdf['producto'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
productosdf
```

```python
resultados=pd.DataFrame()
for productos in productosdf['producto']:
    for tiendas in df_stores['url']:
        resultados=pd.concat([resultados,ue.search_products(mode='avanzada',address="06720 cdmx",producto=productos,url=tiendas),],axis=0)
resultados
```
